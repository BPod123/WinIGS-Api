```python
import matplotlib.pyplot as plt
import os
import pickle
import numpy as np
import pandas as pd
```

```python
os.listdir("Trained_Models/Training1")
```

```python
os.listdir("Trained_Models/Training2")
```

```python
# Data from 1st round of training
recovery1 = []
# (loss, acc, mape, train_time, eval_time, hist, model.name)
for filename in os.listdir("Trained_Models/Training1"):        
    with open('Trained_Models/Training1/{}'.format(filename), 'rb') as f:
        recovery1.append(pickle.load(f))

# Data from 2nd round of training
recovery2 = []
for filename in os.listdir("Trained_Models/Training2"):        
    with open('Trained_Models/Training2/{}'.format(filename), 'rb') as f:
        recovery2.append(pickle.load(f))
```

```python
recovery = [recovery1, recovery2]
```

# Collect the Results in a Table / CSV

```python
training1 = pd.DataFrame(recovery1, columns=['TestLoss', 'TestAccuracy', 'MAPE', 'TrainTime', 'EvalTime', 'Hist', 'Name']).drop(columns="MAPE")
```

```python
training2 = pd.DataFrame(recovery2, columns=['TestLoss', 'TestAccuracy', 'MAPE', 'TrainTime', 'EvalTime', 'Hist', 'Name']).drop(columns="MAPE")
```

```python
training1.set_index("Name")
```

```python
training2.set_index("Name")
```

```python
df_mean = pd.concat((training1.set_index("Name"), training2.set_index("Name")))
```

```python
df_mean.sort_values("TestLoss")
```

```python
mean_index = df_mean.groupby(df_mean.index)
```

```python
df_mean = mean_index.mean()
```

```python
df_mean
```

```python
df_data = df_mean.sort_values("TestLoss")
df_data.to_csv("./Results.csv")
```

```python
df_data.index
```

# Overall Visualisations

```python
# Clean Labels
index = ["AM: LSTM (Simple)",
"A: Attn Dense",
"AM: Dense",
"A: Attn LSTM",
"AM: LSTM2",
"A: Attn BiLSTM",
"AM: LSTM1",
"AM: BiLSTM",
"B: LSTM",
"B: BiLSTM",
"B: Dense",
"C: LSTM (Yang)",
"C: BiLSTM",
"C: LSTM",
"C: Dense",
]
```

```python
color = ["Green" if (name[0] in ["M", "A"]) else "Blue" if name[0] == "B" else "Orange" for name in df_data.index]
```

```python
name_change = {}
for i, old_name in enumerate(df_data.index):
    name_change[old_name] = index[i]
    
print(name_change)
```

```python
plt.rcParams.update({'font.size': 15})
```

```python
index.reverse()
data = list(df_data["TestLoss"])
data.reverse()
color.reverse()
```

```python
# Figure Size
fig, ax = plt.subplots(figsize =(16, 9))
 
# Horizontal Bar Plot
ax.barh(index, data, color=color)
 
# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
 
# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
# Show top values
ax.invert_yaxis()
 
# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 5)),
             fontsize = 12, fontweight ='bold',
             color ='grey')

ax.set_title("Test Losses",
         loc ='left', )
plt.show()
```

```python
# Validation Loss during Training

plt.rcParams.update({'font.size': 22})

fig, axes = plt.subplots(nrows=2, ncols=1)
fig.set_size_inches(20.5, 20.5)

for i, ax in enumerate(axes.flat, start=1):

    ax.set_title(f"Validation Loss during Training {i}")
    ax.set_xlabel("Epochs")
    ax.set_ylabel("Validation Loss")
    for data in recovery[i-1]:
        ax.plot(data[-2]['val_loss'], label=name_change[data[-1]])
    ax.legend(loc='lower right', bbox_to_anchor=(1.25,0))
plt.show()
```

```python
# Validation Loss during Training (Log 10 Scale)

fig, axes = plt.subplots(nrows=2, ncols=1)
fig.set_size_inches(20.5, 20.5)

for i, ax in enumerate(axes.flat, start=1):

    ax.set_title(f"Validation Loss during Training {i}")
    ax.set_xlabel("Epochs")
    ax.set_ylabel("Validation Loss (Log 10 Scale)")
    for data in recovery[i-1]:
        ax.plot(np.log10(data[-2]['val_loss']), label=name_change[data[-1]])
    ax.legend(loc='lower right', bbox_to_anchor=(1.2,0))
plt.show()
```

```python
# Validation Accuracy during Training

fig, axes = plt.subplots(nrows=2, ncols=1)
fig.set_size_inches(20.5, 20.5)

for i, ax in enumerate(axes.flat, start=1):

    ax.set_title(f"Validation Accuracy during Training {i}")
    ax.set_xlabel("Epochs")
    ax.set_ylabel("Validation Accuracy")
    for data in recovery[i-1]:
        ax.plot(data[-2]['val_accuracy'], label=name_change[data[-1]])
    ax.legend(loc='lower right', bbox_to_anchor=(1.2,0))
plt.show()
```

# Visualisations for LSTM Models

```python
# Validation Loss during Training

plt.rcParams.update({'font.size': 22})

fig, axes = plt.subplots(nrows=2, ncols=1)
fig.set_size_inches(20.5, 20.5)

for i, ax in enumerate(axes.flat, start=1):

    ax.set_title(f"Validation Loss during Training {i}")
    ax.set_xlabel("Epochs")
    ax.set_ylabel("Validation Loss")
    for data in recovery[i-1]:
        name = data[-1]
        if name.split("_")[1][:4] == "LSTM":
            ax.plot(data[-2]['val_loss'], label=name_change[name])
    ax.legend(loc='lower left', bbox_to_anchor=(0.75,0.35))
plt.show()
```
